# 의사결정트리 (Decision Tree)

* 스무고개 퀴즈/ 아키네이터 로 이해하자

* 복잡도가 높다 = 엔트로피가 높다.
* 엔트로피 : 불확실성을 수치적으로 표현한 값
* 낮춰야 좋은 데이터

* 정보 획득 (IG,Information Gain)
* 정보 획득량 = 사전 엔트로피(불확실성) - 사후 엔트로피 (불확실성)
* 한쪽의 클래스만 데이터가 있는게 엔트로피 0 

* 지니 계수

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
iris.target
```

```python
labels = pd.DataFrame(iris.target)
data = pd.DataFrame(iris.data)
data
```

```python
labels.columns = ["labels"]
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
```

```python
labels
data
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.concat([data,labels],axis=1)
data
```

```python
data[['Sepal length','Sepal width']]
```

```python
feature = data[["Sepal length","Sepal width"]]
feature
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

* K-평균 사용

```python
model = KMeans(n_clusters=3)  # 군집 3개 설정
```

```python
model.fit(feature)
```

```python
pred = pd.DataFrame(model.predict(feature))
pred.columns = ["predict"]
res = pd.concat([feature,pred],axis=1)
res
```

```python
data
```

```python
centers = pd.DataFrame(model.cluster_centers_,columns=["Sepal length","Sepal width"]) # 중심점
```

```python
c_x = centers["Sepal length"]
c_y = centers["Sepal width"]
```

```python
plt.scatter(c_x,c_y)
plt.show()
```

```python
plt.scatter(res["Sepal length"],res["Sepal width"], c = res["predict"])
plt.scatter(c_x,c_y , c = "red", marker = "D")
plt.show()
```

```python
pd.crosstab(data["labels"],res["predict"])
```

```python
q1 = data["age"].quantile(0.25) # 16.504
q3 = data["age"].quantile(0.75) # 18.391
```

```python
iqr = q3-q1
iqr
```

```python
q1-1.5*iqr # 13.672 Lower Fence
q3+1.5*iqr # 21.222 Upper Fence
# 13.672>x or 21.222<x Outliers
df = data[(data["age"]>(q1-1.5*iqr))&(data["age"]<(q3+1.5*iqr))]
df
# 아웃라이어를 제외한 정상 범위의 테이터만 추출하여 데이터 프레임 구성
```

```python
def gen_to_num(x):
    if x =="M":
        return 1
    if x =="F":
        return 2
    if x =="not disclosed":
        return 3
```

```python
data.head()
data.groupby(["cluster"]).groups
data.groupby(["cluster"]).count()["age"]
```

```python
wcss = [] # within cluster sum of square
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++' ,max_iter=300,random_state=0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss) 
plt.title('THe Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel("WCSS")
plt.show() # Elbow 그래프
```
