```python
def newSeason(month):
    if month in [12,1,2]:
        return 4
    elif  month in [3,4,5]:
        return 1
    elif  month in [6,7,8]:
        return 2
    elif  month in [9,10,11]:
        return 3
```

```python
train.month
train.season
#봄(3,4,5월):1...겨울:4(12,1,2월)
train['season']=train.month.apply(newSeason)
```
